```python
#연습 문제
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
print(df)
```

```python
#나이에 따라 직원을 분류하는 함수 정의 
#young, mid, experience
def Age_category(age):
    if age < 30:
        return 'young'
    elif age <40:
        return 'mid'
    else:
        return 'exprienced'
```

```python
#데이터프레임에 나이(Age)에 따라 직원 분류 함수 적용해서 'Category' 컬럼을 추가
df['Category'] = df['Age'].apply(Age_category)
df
```

```python
# salary에 보너스 10% 를 증가한 컬럼 'Bones'를 추가 > 람다함수 사용
Bonus = lambda x: x*0.10 
df['Bones'] = df['Salary'].apply(Bonus)
df
```

```python
def my_zscore(x): #표준점수, 나의 값이 전체 데이터 평균과 비교해서 얼마나 떨어져 있는지 보여주는 함수
    return ((x -x.mean())/x.std())
```
